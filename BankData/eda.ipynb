```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV into a pandas dataframe
df = pd.read_csv('BankChurners.csv')
```

## Dimensions

```python
# Show the dimensions in the dataframe (Rows, Columns)
rows, columns = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

## Summary Statistics

```python
# Display floats to 4dp
pd.options.display.float_format = '{:.4f}'.format
# Tranpose the dataframe to show data efficiently
df.describe().transpose()
```

## Correlation

```python
# Using standard correlation coefficient. "Pearson" method
df.corr()
```

```python
# Using standard correlation coefficient. "Pearson" method
df.corr(method="kendall")
```

```python
# Using Kendall Tau correlation coefficient. "Kendall" method
df.corr(method="kendall")
```

```python
# Using Spearman rank correlation. Method "spearman"
df.corr(method="spearman")
```

## Data Visualization

```python
# Check for unique values for all columns. Help to determine what to visualize.

# Get all columns
all_columns = df.columns
# Loop though all columns
for one in all_columns:
    print(f"{one} - Column")
    print(df[one].unique())
```

```python
# A relation of the marital status and customer ages
sns.displot(data=df, x="Marital_Status", kind="hist", y="Customer_Age")
plt.title("A relation of marital status and customer age")
plt.xlabel("Marital Status")
plt.ylabel("Customer Age")
plt.xticks(rotation=45)
plt.show()
```

```python
# A relation of income category, customer ages and genders
sns.stripplot(data=df, x="Income_Category", y="Customer_Age", hue="Gender")
plt.title("A relation of income category, customer age and gender")
plt.xlabel("Income Category ($)")
plt.ylabel("Customer Age")
plt.xticks(rotation=45)
plt.show()
```

```python
# A relation of attrition flag and card category
sns.catplot(data=df, kind="count", x="Card_Category", hue="Attrition_Flag")
plt.title("A relation of attrition flag and card category")
plt.xlabel("Card Category")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# A relation of education level, age and gender
sns.stripplot(data=df, x="Customer_Age", y="Education_Level", hue="Gender")
plt.title("A relation of education level, age and gender")
plt.xlabel("Customer Age")
plt.ylabel("Education Level")
plt.show()
```

```python
# A relation of attrition flag and months inactive
sns.histplot(data=df, x="Months_Inactive_12_mon", hue="Attrition_Flag")
plt.title("A relation of attrition flag and months inactive")
plt.xlabel("Months Inactive In A Year")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# A relation of card category and credit limit
sns.histplot(data=df, x="Credit_Limit", kde=True, hue="Card_Category")
plt.title("A relation of card category and credit limit")
plt.xlabel("Credit Limit")
plt.ylabel("Number of Customers")
plt.show()
```
